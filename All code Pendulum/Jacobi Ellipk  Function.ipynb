```python
import numpy as np
from scipy.special import ellipk
import matplotlib.pyplot as plt

# Define theta_0 as an array of initial angles
theta_0 = np.linspace(0, np.pi, 100)

# Calculate m for each theta_0
m = np.sin(theta_0 / 2) ** 2

# Calculate T and omega for each value of m
T = 4 * ellipk(m)  # Period T for each m
omega = 2 * np.pi / T  # Angular frequency omega for each T

# Plot the results
plt.plot(theta_0, omega, label=r'$\omega$ vs $\theta_0$', color='b')
plt.xlabel(r'Initial Angle $\theta_0$ (radians)')
plt.ylabel(r'Angular Frequency $\omega$ (radians/second)')
plt.title('Angular Frequency as a Function of Initial Angle for a Nonlinear Pendulum')
plt.legend()
plt.grid()
plt.show()
```
