```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the pendulum function for the small angle approximation
def pendulum_small_angle(t, state):
    theta, omega = state
    dtheta_dt = omega
    domega_dt = -theta  # Small angle approximation
    return [dtheta_dt, domega_dt]

# Define the pendulum function for the Taylor series approximation
def pendulum_taylor(t, state):
    theta, omega = state
    dtheta_dt = omega
    domega_dt = -(theta - (theta**3 / 6))  # Taylor series approximation
    return [dtheta_dt, domega_dt]

# Define the pendulum function for the full nonlinear dynamics
def pendulum_nonlinear(t, state):
    theta, omega = state
    dtheta_dt = omega
    domega_dt = -np.sin(theta)  # Nonlinear dynamics
    return [dtheta_dt, domega_dt]

# Time span for the simulation
t_span = (0, 10)  # From t=0 to t=10 seconds
t_eval = np.linspace(0, 10, 1000)  # Time points at which to store the result

# Range of initial angles
theta_0_values = np.linspace(0, np.pi, 100)
omega_values_small_angle = []
omega_values_taylor = []
omega_values_nonlinear = []

# Loop over each initial angle for the small angle approximation
for theta_0 in theta_0_values:
    state = [theta_0, 0]  # Initial angular velocity is 0
    sol = solve_ivp(pendulum_small_angle, t_span, state, t_eval=t_eval)
    
    # Extract theta values and find zero crossings
    theta_values = sol.y[0]
    zero_crossings = np.where(np.diff(np.sign(theta_values)))[0]
    
    # Calculate the period if there are enough zero crossings
    if len(zero_crossings) > 1:
        T = sol.t[zero_crossings[1]] - sol.t[zero_crossings[0]]  # First full period
        omega = 2 * np.pi / T
        omega_values_small_angle.append(omega)
    else:
        omega_values_small_angle.append(np.nan)

# Loop over each initial angle for the Taylor series approximation
for theta_0 in theta_0_values:
    state = [theta_0, 0]
    sol = solve_ivp(pendulum_taylor, t_span, state, t_eval=t_eval)
    
    theta_values = sol.y[0]
    zero_crossings = np.where(np.diff(np.sign(theta_values)))[0]
    
    if len(zero_crossings) > 1:
        T = sol.t[zero_crossings[1]] - sol.t[zero_crossings[0]]
        omega = 2 * np.pi / T
        omega_values_taylor.append(omega)
    else:
        omega_values_taylor.append(np.nan)

# Loop over each initial angle for the full nonlinear dynamics
for theta_0 in theta_0_values:
    state = [theta_0, 0]
    sol = solve_ivp(pendulum_nonlinear, t_span, state, t_eval=t_eval)
    
    theta_values = sol.y[0]
    zero_crossings = np.where(np.diff(np.sign(theta_values)))[0]
    
    if len(zero_crossings) > 1:
        T = sol.t[zero_crossings[1]] - sol.t[zero_crossings[0]]
        omega = 2 * np.pi / T
        omega_values_nonlinear.append(omega)
    else:
        omega_values_nonlinear.append(np.nan)

# Plotting results
plt.figure(figsize=(10, 5))
plt.plot(omega_values_small_angle, theta_0_values, label='Small Angle Approximation ($d\omega/dt = -\\theta$)', color='b')
plt.plot(omega_values_taylor, theta_0_values, label='Taylor Series Approximation ($d\omega/dt = -(\\theta - \\theta^3/6)$)', color='r')
plt.plot(omega_values_nonlinear, theta_0_values, label='Nonlinear Dynamics ($d\omega/dt = -\\sin(\\theta)$)', color='g')
plt.xlabel(r'Angular Frequency $\omega$ (radians/second)')
plt.ylabel(r'Initial Angle $\theta_0$ (radians)')
plt.title('Initial Angle as a Function of Angular Frequency for a Pendulum')
plt.legend()
plt.grid()
plt.show()
```
