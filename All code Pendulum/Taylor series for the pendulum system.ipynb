```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define function for the system of ODEs (pendulum dynamics with second-order Taylor series approximation)
def pendulum_taylor(t, state):
    theta, omega = state
    dtheta_dt = omega
    domega_dt = -(theta - theta**3 / 6)  # Taylor series approximation
    return [dtheta_dt, domega_dt]

# Initial conditions: theta(0) = 0.1
theta_0 = 0.005   # Initial angle in radians
omega_0 = 0.0   # Initial angular velocity
state = [theta_0, omega_0]  # Initial state [theta(0), omega(0)]

# Time span for the simulation
t_span = (0, 10)  # From t=0 to t=10 seconds
t_eval = np.linspace(0, 10, 1000)  # Time points at which to store the result

# Solve the system using solve_ivp (Runge-Kutta method)
sol = solve_ivp(pendulum_taylor, t_span, state, t_eval=t_eval)

# Plot the results
plt.figure(figsize=(10, 5))
plt.plot(sol.t, sol.y[0], label='Theta (angle)')
plt.plot(sol.t, sol.y[1], label='Omega (angular velocity)')
plt.xlabel('Time (s)')
plt.ylabel('State (angle and velocity)')
plt.title('Pendulum Dynamics with Taylor Series Approximation (2nd Order)')
plt.legend()
plt.grid()
plt.show()
```
