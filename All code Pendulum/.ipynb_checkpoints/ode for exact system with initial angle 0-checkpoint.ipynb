```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#define function of ode
def pendulum(t,state):
    theta, omega = state
    dtheta_dt = omega
    domega_dt = -np.sin(theta)
    return[dtheta_dt,domega_dt]

# Initial conditions: theta(0) and omega(0)
theta_0=0  # Initial angle (0 degree)
omega_0=0   ## Initial angular velocity
state=[0,0]

#time for plotting 
t_span=(0,10)
t_eval=np.linspace(0,10,1000)

#solution using solve_ivp
sol=solve_ivp(pendulum,t_span,state,t_eval=t_eval)


#plot the result
plt.plot(sol.t,sol.y[0],label='Theta (angle)')
plt.plot(sol.t,sol.y[1],label='omega (Angular velocity) ')
plt.xlabel('Time(s)')
plt.ylabel('state(angle and velocity)')
plt.title('pendulum dynamics')
plt.legend()
plt.grid()
plt.show()
```
